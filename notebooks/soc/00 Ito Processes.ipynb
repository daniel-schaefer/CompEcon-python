Simulate geometric Brownian motion

```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn')
```

## Model Parameters

```python
T = 1
n = 365
t = np.linspace(0, T, n)
h = t[1] - t[0]

ùúá = 0.1
ùúé = 0.05
```

## Simulate

```python
m = 3
z = np.random.randn(n,m)
s = np.zeros((n,m))
s[0] = 1
for i in range(n-1):
    s[i+1] = s[i] + ùúá*s[i]*h + ùúé*s[i]*np.sqrt(h)*z[i]
```

## Plot

```python
fig, ax = plt.subplots(figsize=[8,4])
ax.plot(t,s)
ax.set(xlabel='$t$',
       ylabel='$s(t)$',
       title='Simulated Geometric Brownian Motion, $\mu=0.1$, $\sigma=0.05$');
```
